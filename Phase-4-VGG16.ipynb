# PREPARING DATSET FOR USING IMAGE DATA GENERATOR PAKAGE FOR TRANFER LEARNING.

```python
from __future__ import print_function
import pandas as pd
import shutil
import os
import sys

#labels = pd.read_csv(r'D:\4 th semester\Projects\DSP\train.zip.001\editedtrainLabels.csv')
path = './'
labels_csv_path = path + 'labels.csv'
# sample_submission_csv_path = path + 'sample_submission.csv'
# submission_csv_path =  '../input/sample_submission.csv'
train_path = path + 'train'
test_path = path + 'test'
labels = pd.read_csv(labels_csv_path)

train_dir = train_path
DR = r"./DR/"


if not os.path.exists(DR):
    os.mkdir(DR)

for filename, class_name in labels.values:
    # Create subdirectory with `class_name`
    if not os.path.exists(DR + str(class_name)):
        os.mkdir(DR + str(class_name))
    src_path = train_dir + '/'+ filename + '.jpg'
    dst_path = DR + str(class_name) + '/' + filename + '.jpg'
    try:
        shutil.copy(src_path, dst_path)
        print("sucessful")
    except IOError as e:
        print('Unable to copy file {} to {}'
              .format(src_path, dst_path))
    except:
        print('When try copy file {} to {}, unexpected error: {}'
              .format(src_path, dst_path, sys.exc_info()))
```
